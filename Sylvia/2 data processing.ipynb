```python
import pandas as pd
import numpy as np
import pickle
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
full_dataset= pd.read_csv("0320_diabetes_clean.csv")
```

```python
full_dataset.head()
```

# remove column payer_code
since it contains a large number of missing data, and we believe it does not effect the response (readmission) 

```python
full_dataset.drop(['payer_code'], axis = 1, inplace = True)
```

```python
full_dataset.shape
```

### Based on the paper, remove " dischargeDisposition"="discharge to hospice" 
* 396 rows + 368 rows dropped

```python
full_dataset.drop(full_dataset[full_dataset['dischargeDisposition']=="Hospice / home"].index,axis=0,inplace=True)

full_dataset.drop(full_dataset[full_dataset['dischargeDisposition']=="Hospice / medical facility" ].index,axis=0,inplace=True)
```

```python
full_dataset.shape
```

### drop "encounter_id" and "patient_nbr" column
* these are just  unique IDs for the patients

```python
full_dataset.drop(['encounter_id'], axis = 1, inplace = True)
full_dataset.drop(['patient_nbr'], axis = 1, inplace = True)
```

### drop rows  "race" =="?"
* 2152 rows dropped

```python
full_dataset.drop(full_dataset[full_dataset['race']=="?"].index,axis=0,inplace=True)
```

```python
full_dataset.shape
```

```python
full_dataset.to_csv("0406_diabetes_clean.csv")
```
